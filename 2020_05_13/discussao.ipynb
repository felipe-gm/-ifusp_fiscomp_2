Related third party imports.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Variaveis globais

```python
mu, sigma = .0, [.25,.5,1.,2.,4.] # mean and standard deviation
```

Constantes

```python
T = 10**3   # Numero de passos da simulacao
M = int(1e4)     # Numero de caminhadas independentes
```

Draw samples from the distribution

```python
s = np.random.default_rng().normal(mu, sigma, (2,T,M,len(sigma)))
```

trajetoria

```python
caminhada = np.cumsum(s,axis=1)      # sum over columns for each of s
```

quadrado das distancias

```python
sqrd_dist = caminhada[0]**2+caminhada[1]**2
```

plotar medias

```python
plt.plot(np.mean(sqrd_dist, axis=1))
```

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(np.mean(sqrd_dist, axis=1))
```

Modificações na variância mostram que a distáncia quadrática média em
função do tempo são funções lineares com diferentes coeficientes 
angulares.
$$ R(t) \propto \sigma \sqrt{t} $$
